# Matplotlib

- Plot 1 : simple plot

```python
import matplotlib.pyplot as plt
x = ["Javascript", "Python", "Java", "C++"]
y = [70, 97, 50, 27]
plt.title("Programming Languages Sats 2024")
plt.bar(x,y)
plt.show()
```

- Plot 2 : Modified
    - color property

```python
x = ["Javascript", "Python", "Java", "C++"]
y = [70, 95, 60, 20]
c = ["blue", "green", "orange", "red"]
plt.title("Programming Languages Sats 2024")
plt.bar(x,y, color=c)
# plt.barh(x, y, color=c)
plt.show()
```

- Modified
    - align
    - width
    - linewidth
    - edgecolor

```python
x = ["Javascript", "Python", "Java", "C++"]
y = [70,90,50,20]
c = ["blue", "green", "brown", "red"]
plt.title("Programming Language Stats 2024")

plt.bar(x,y, color=c, align='edge', width=0.4, linewidth=2.4,  edgecolor="black")
plt.show()
```

- Horizontal Bar

```python
plt.title("Programming Languages Stats 2024")
plt.barh(x,y, color='green')
```
