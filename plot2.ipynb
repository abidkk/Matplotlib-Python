# Line Graph

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [100, 50, 70, 40]

plt.plot(x,y)
plt.show()
```

- Plotting without line

```python
plt.plot(x,y, 'o')
plt.show()
```

```python
s = [10,50,90,20,60,30,80]
plt.plot(s)
plt.show()
```

- Marker

```python
s = [10,50,90,20,60,30,80]
plt.plot(s, marker='o')
plt.show()
```

```python
s = [10,50,90,20,60,30,80]
plt.plot(s, marker='*')
plt.show()
```

```python
plt.plot(s, marker='D')
```

```python
plt.plot(s, marker='H')
```

- Marker Properties
    - ms
    - mec
    - mfc

```python
plt.plot(s, marker='H', ms=20, mec='black', mfc='gray')
```

- line Style
    - dotted
    - dashed
    - dashdot
    

- Dotted

```python
plt.plot(s, linestyle='dotted')
```

- Dashed

```python
plt.plot(s, linestyle='dashed')
```

- Dashdot

```python
plt.plot(s, linestyle='dashdot')
```

- Line color and Width

```python
plt.plot(s, linewidth=7.5, color='green')
```

# Multi Line Plot

```python
x = [40,80,50,20,70]
y = [65,35,95,55,15]
plt.plot(x, color='r')
plt.plot(y, color='g')
plt.show()
```

```python
x = [40,80,50,20,70]
y = [65,35,95,55,15]
z = [85,55,20,45,25]
plt.plot(x, color='r')
plt.plot(y, color='g')
plt.plot(z, color='b')
plt.show()
```

- Plot Title and Labels

```python
x = [40,80,50,20,70]
y = [65,35,95,55,15]
z = [85,55,20,45,25]
year = [2003, 2005, 2010, 2024]
plt.plot(x, color='r', label='Java')
plt.plot(y, color='g', label='Python')
plt.plot(z, color='b', label='Javascript')
plt.title("Programming Language Stats 2024")
plt.legend()
# plt.xlabel("Index")  
# plt.ylabel("Popularity Score")
plt.show()
```

# Grid to the Plot

```python
x = [40,80,50,20,70]
plt.title("Programming Language")
plt.plot(x, label="Python")
plt.legend()
plt.grid()
plt.show()
```

- Grid : x-axis

```python
x = [40,80,50,20,70]
plt.title("Programming Language")
plt.plot(x, label="Python")
plt.legend()
plt.grid(axis='x')
plt.show()
```

- Grid : y-axis

```python
x = [40,80,50,20,70]
plt.title("Programming Language")
plt.plot(x, label="Python")
plt.legend()
plt.grid(axis='y')
plt.show()
```
