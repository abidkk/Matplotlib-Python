# Line Graph

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [100, 50, 70, 40]

plt.plot(x,y)
plt.show()
```

- Plotting without line

```python
plt.plot(x,y, 'o')
plt.show()
```

```python
s = [10,50,90,20,60,30,80]
plt.plot(s)
plt.show()
```

- Marker

```python
s = [10,50,90,20,60,30,80]
plt.plot(s, marker='o')
plt.show()
```

```python
s = [10,50,90,20,60,30,80]
plt.plot(s, marker='*')
plt.show()
```

```python
plt.plot(s, marker='D')
```

```python
plt.plot(s, marker='H')
```

- Marker Properties
    - ms
    - mec
    - mfc

```python
plt.plot(s, marker='H', ms=20, mec='r', mfc='black')
```
