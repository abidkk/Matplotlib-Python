# Subplot

- Multiple Plots

```python
import matplotlib.pyplot as plt
```

```python
x = [20,60,30,50,40]
y = [70,30,80,50,90]

# Plot 1
plt.subplot(1,2,1)
plt.plot(x)

# Plot 2
plt.subplot(1,2,2)
plt.plot(y)
plt.show()
```

- Horizontal Plots

```python
# Plot 1
plt.subplot(2,1,1)
plt.plot(x)

# Plot 2
plt.subplot(2,1,2)
plt.plot(y)
plt.show()
```

```python
x = [10,30,60,20,80]
y = [30,20,80,40,60]
z = [40,30,20,10,90]

# Plot 1
plt.subplot(2,2,1) # row, col, index
plt.plot(x)
plt.title("Java")


# Plot 2
plt.subplot(2,2,2)
plt.plot(y)
plt.title("C++")

# Plot 2
plt.subplot(2,2,3)
plt.plot(z)
plt.title("Python")

# auto adjust gap b.w subplots
# plt.tight_layout() 

# Adjust space between subplots Manually
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()


```
