# Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x = ['Mon', "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
y = [5,6,2,8,10,7,3]
plt.scatter(x,y)
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day")
plt.ylabel("Range")

plt.show()
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
studetns = [300, 240, 200, 130, 290, 250, 180]
plt.scatter(days, studetns)


water = [100, 80, 70, 40, 95, 70, 60]
plt.scatter(days, water)
```
